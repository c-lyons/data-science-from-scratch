```python
from sklearn.datasets import make_blobs
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
```

```python
X, y = make_blobs(n_samples=500, centers=3, n_features=2, cluster_std=1.5)
```

```python
plt.scatter(X[:, 1], X[:, 0], c=y )
```

```python
K = 3
y_pred = np.random.choice(K, size=len(y), replace=True)
means = [(X[y_pred==i][:,0].mean(), X[y_pred==i][:,1].mean()) for i in range(K)]
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
for i in range(K):
    plt.scatter(means[i][0], means[i][1], c='r')
```

```python
# Keep running this to watch iterative progress - REPLACE WITH FOR LOOP IN PRACTISE
for i in range(len(X)):
    dist_to_k = []
    for k in range(K):
        dist_to_k.append(sqrt((X[i][0] - means[k][0])**2 + (X[i][1] - means[k][1])**2))
    y_pred[i] = dist_to_k.index(min(dist_to_k))  # assigning new y labels closest to centroids
means = [(X[y_pred==i][:,0].mean(), X[y_pred==i][:,1].mean()) for i in range(K)]  # getting new centroids for next iter
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
for i in range(K):
    plt.scatter(means[i][0], means[i][1], c='r')
```
